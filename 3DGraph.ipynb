### 3D surface plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx,yy = np.meshgrid(x,y)
xx.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis') #use color="" or cmap='viridis..
fig.colorbar(p) # show line of color
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis') #use color="" or cmap='viridis..
fig.colorbar(p)
```

## Contour plots

make 3D graph into 2D graph

```python
z = xx**2 + yy**2

fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='viridis') # use contour
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis') # use contourf
fig.colorbar(p)

# 3D graph shown in 2D dark side is down side while lighter is upper side
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis') 
fig.colorbar(p)

# blue is bottom , yellow is top 
```

## Heat map

```python
# plot grid graph colums show ball row show over 
# make senario like 19 over ki 4 ball pr kitne 6 lge h

delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
# we work on ball from 1-6 only and only ball who goes for 6
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run'] == 6)]
temp_df.head()
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)

plt.yticks(delivery['overs'].unique(),list(range(1,21))) # show all over no
plt.xticks(np.arange(0,6),list(range(1,7))) # show all ball no
plt.colorbar()
```
