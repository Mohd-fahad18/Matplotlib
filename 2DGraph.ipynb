## 2D Line Graph 

### Bivariate

CATEGORICAL -> numerical and numerical

USE CASE -> Time series data

```python
# Ploting simple function 
price = [48000,54000,57000,49000,47000,45000]  #numerical
year = [2015,2016,2017,2018,2019,2020]   #categorical 

plt.plot(year,price)   # (X axis -> Year) , (Y axis -> price)
```

```python
# from pandas dataframe 
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# ploting multiple plot

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# label title

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma VS Virat Kohli Carrer comparison')
plt.xlabel('Season')
plt.ylabel('Run Score')
```

```python
# Color(hex) and line (width and style ) and marker (size)

plt.plot(batsman['index'],batsman['V Kohli'],color='#020dbb',linewidth=4)  # hex color code
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed') #( dashed),(dotted),( dash dot) are line type

plt.title('Rohit Sharma VS Virat Kohli Carrer comparison')
plt.xlabel('Season')
plt.ylabel('Run Score')
```

```python
# marker = 'D','+','o','<','>'..(show on points)

plt.plot(batsman['index'],batsman['V Kohli'],color='#020dbb',marker='D',markersize=8) 
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='>') 
plt.title('Rohit Sharma VS Virat Kohli Carrer comparison')
plt.xlabel('Season')
plt.ylabel('Run Score')
```

```python
# Give name to line 
# legend -> location

plt.plot(batsman['index'],batsman['V Kohli'],color='#020dbb',marker='d',label='Virat')  
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='o',label='Rohit')
plt.title('Rohit Sharma VS Virat Kohli Carrer comparison')
plt.xlabel('Season')
plt.ylabel('Run Score')

plt.legend(loc='upper left') #upper right , right ....
```

```python
# limiting axes

price = [48000,54000,57000,49000,47000,45000,450000]  
year = [2015,2016,2017,2018,2019,2020,2021]   

plt.plot(year,price)
plt.ylim(0,75000) # give range according to us 
#plt.xlim(2017,2019)
plt.grid()
plt.show() # must have to write to show graph
```

# Bar Chart

Numercal Vs Categorical

```python
# Simple ba chart 

children=[10,20,40,10,30]
color =['red','blue','green','yellow','pink']

plt.bar(color,children,color='red')
plt.grid()
```

```python
# Horizontal graph ( barh )
plt.barh(color,children,color='red')
```

```python
# color and label 
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
# problem multiple graph 
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2015'])
```

```python
# use width to solve problem

plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='green')

plt.xticks(np.arange(df.shape[0]),df['batsman']) # must write to show name
```

```python
# problem ( overlapping )

children=[10,20,40,10,30]
color =['red red red ','blue blue blue ','green green green','yellow yellow','pink pink pink']

plt.bar(color,children,color='black')
plt.xticks(rotation='vertical') # solve through this
```

```python
df
```

```python
# stacked bar chart ( add previous graph by bottom = df(name of file['columns name in file']))

plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2015'] + df['2016']),label='2017')

plt.legend()
plt.grid()
plt.show()
```

## Pie Chart

univariate/bivariate Analysis (single or multiple col)

Categorical Vs Numerical

USE CASE: To find contribution on a standard scale

```python
# simple Data 
data=[23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)
plt.show()
```

```python
# Dataset 
df = pd.read_csv('gayle-175.csv')
df
```

```python
# percentage and color

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','brown','red','yellow','cyan'])
plt.show
```

```python
# make out perticular data (explode shadow )

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0.2],shadow=True)
plt.show
```
